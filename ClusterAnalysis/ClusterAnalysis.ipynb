```python
from sklearn.preprocessing import StandardScaler

def preprocess_data(df):
    # Handle missing values (if any)
    df.fillna(df.mean(), inplace=True)

    # Normalize/Scale the data (important for KMeans)
    scaler = StandardScaler()
    scaled_data = scaler.fit_transform(df)

    return scaled_data
```

```python
from sklearn.cluster import KMeans
import numpy as np

def perform_clustering(data, n_clusters=3):
    kmeans = KMeans(n_clusters=n_clusters, random_state=42)
    labels = kmeans.fit_predict(data)
    centers = kmeans.cluster_centers_

    return labels, centers
```

```python
import matplotlib.pyplot as plt
from sklearn.decomposition import PCA

def plot_clusters(data, labels, centers):
    # Reduce dimensions to 2D for visualization
    pca = PCA(n_components=2)
    data_2d = pca.fit_transform(data)

    # Plot the clusters
    plt.figure(figsize=(8, 6))
    plt.scatter(data_2d[:, 0], data_2d[:, 1], c=labels, cmap='viridis', label="Clusters")

    # Plot the cluster centers
    plt.scatter(centers[:, 0], centers[:, 1], s=300, c='red', marker='x', label="Centroids")
    plt.title("KMeans Clustering")
    plt.legend()
    plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.datasets import load_iris
from sklearn.cluster import KMeans
```

```python
def main():
    # Load Iris dataset (as an example)
    iris = load_iris()
    data = pd.DataFrame(iris.data, columns=iris.feature_names)

    # Preprocess the data
    data = preprocess_data(data)

    # Perform clustering using KMeans
    n_clusters = 3  # As Iris has 3 classes
    labels, centers = perform_clustering(data, n_clusters)

    # Visualize the clusters
    plot_clusters(data, labels, centers)

if __name__ == "__main__":
    main()
```
